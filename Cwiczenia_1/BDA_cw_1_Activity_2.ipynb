```python
# 1. Import pandas and numpy library.

import numpy as np
import pandas as pd
```

```python
# 2. Read the RadNet dataset from the U.S. Environmental Protection Agency.

df = pd.read_csv("RadNet_Laboratory_Analysis.csv")
```

```python
# 3. Create a list with numeric columns for radionuclides in the RadNet dataset.

# df.head()
numeric = df.iloc[:,6:]
numeric_list = list()
for i in numeric:
    numeric_list.append(i)
print(*numeric_list, sep = ", ")    
```

```python
# 4. Use the apply method on one column, with a lambda function that compares the Non-detect
# string

df['Ba-140'] = df['Ba-140'].apply(lambda x: np.nan if x == "Non-detect" else x)
```

```python
# 5. Use applymap method to replace text values "Non-detect" with .np.nan in all numeric
# columns at once, do the same with "ND" value

df.iloc[:, :] = df.iloc[:, :].applymap(lambda x: np.nan if x == 'Non-detect' else x)
df.iloc[:, :] = df.loc[:, :].applymap(lambda x: np.nan if x == 'ND' else x)

df.head(10)
```

```python
# 6. Use the same applymap method to remove all spaces (from begining and end) in all other
# columns (non-numeric). Use strip() method to remove spaces.

# df.dtypes

not_numeric = df.iloc[:,:6]
not_numeric_list = list()
for i in not_numeric:
    not_numeric_list.append(i)
print(*not_numeric_list, sep = ", ")

for i in not_numeric_list:
    df[i]=df[i].astype('string')

df.iloc[:, :6] = df.iloc[:, :6].apply(lambda x: x.strip() if x.dtype=='String' else x)
```

```python
# 7. Change datatypes of columns using methods (use df.dtypes() to check datatypes):

df.dtypes

# pd.to_datetime()
# pd.to_numeric()

date_type = df.iloc[:,2:4]
date_type_list = list()
for i in date_type:
    date_type_list.append(i)
    
for i in date_type_list:
    df[i] = pd.to_datetime(df[i])

for j in numeric_list:
    df[j] = pd.to_numeric(df[j])

df.dtypes
```
